<h1>HOUSE PRICE PREDICTION</h1>

<h2>Import Libraries:</h2>

```python
#Import Numpy to perform mathematical computations over nd-arrays
import numpy as np

#Import Pandas to prepare the data for analysis
import pandas as pd
#Import is_string_dtype() method to check if an attribute / feature consists of string values or not
from pandas.api.types import is_string_dtype

#Import XGBoost to train the model
from xgboost import XGBRegressor

#Import train_test_split() method from Sk-learn to split the dataframe into training and validation sets
from sklearn.model_selection import train_test_split, RandomizedSearchCV, StratifiedKFold
from scipy.stats import randint, uniform

#Import Matplotlib for visualisation of Insights
import matplotlib.pyplot as plt
#Import Seaborn for data visualisation
import seaborn as sns
sns.set()

#Ignore the warnings
import warnings
warnings.filterwarnings('ignore')

#Set the random seed to a particular state
seed=50
np.random.seed(seed)

#The svg format provides sharp visualisations
%config InlineBackend.figure_format = 'svg'
```

<h2>Extract Training & Test Data:</h2>

```python
#Import data into a data frame
data = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv')
#Import test data into a data frame
test = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')
actual_y = pd.read_csv('../input/house-prices-advanced-regression-techniques/sample_submission.csv').SalePrice
#Save Id(s) of test data for later submission
Id = test.Id

#Select columns for X
cols = data.columns
#Drop the target label (SalePrice)
cols=cols.drop('SalePrice')

data.info()
```

```python
#Drop down data instances whose Sale Price is not defined
data = data[data.SalePrice.notnull()]

#Drop irrelevant attributes
cols=cols.drop(['Alley','PoolQC','MiscFeature'])

#Display 1st five instances of data
data.head()
```

<h2>Data Exploration, Cleaning and Transformation:</h2>

```python
#Seperate categorical from numerical data
categorical = data.select_dtypes(exclude = [np.number])
numerical = data.select_dtypes(include = [np.number])
```

```python
%config InlineBackend.figure_format = 'png'
#Plot numerical data against Sale Price
for idx, feat in enumerate(numerical.columns.difference(['SalePrice'])):
    ax = sns.jointplot(x=feat, y='SalePrice', data=numerical, kind='scatter');
    ax.set_axis_labels(feat, 'SalePrice');
plt.show()
```

```python
#Removal of Outlier
data=data.drop(data[data['1stFlrSF']>4000].index)
```

```python
#Describe Sale Price's characteristics
data['SalePrice'].describe()
```

```python
%config InlineBackend.figure_format = 'svg'
#Plot the SalePrice of each instance
g = sns.distplot(data['SalePrice'])
for item in g.get_xticklabels():
    item.set_rotation(90)
plt.show()
```

```python
#skewness and kurtosis
print("Skewness: %f" % data['SalePrice'].skew())
print("Kurtosis: %f" % data['SalePrice'].kurt())
```

```python
#Tranform categorical data into numerical data for training purpose
for col, col_data in data.items():
    if is_string_dtype(col_data):
        data[col] = data[col].astype('category').cat.as_ordered().cat.codes
        test[col] = test[col].astype('category').cat.as_ordered().cat.codes
```

```python
y = data.SalePrice
X = data[cols]

test=test[cols]
```

<h2>Train Model:</h2>

```python
params_var = {
    'max_depth': [1, 2, 3, 4, 5],
    'gamma': [0, 0.5, 1],
    'n_estimators': randint(1, 1001), # uniform discrete random distribution
    'learning_rate': uniform(), # gaussian distribution
    'subsample': uniform(), # gaussian distribution
    'colsample_bytree': uniform() # gaussian distribution
}

params_fixed = {
    'silent': 1
}

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
```

```python
#Train a model using XGBRegressor
model = RandomizedSearchCV(
    estimator=XGBRegressor(**params_fixed, seed=seed),
    param_distributions=params_var,
    n_iter=10,
    cv=cv,
    scoring='r2',
    random_state=seed
)
#Fit the model over training data
model.fit(X, y, eval_set=[(test, actual_y)], verbose=False)
```

```python
print("Best Score:",model.best_score_)
print("Best Parameters: ",model.best_params_)
```

```python
model.best_estimator_
```

<h2>Test Data:</h2>

```python
#Predict Sale Price for houses in test data
predicted_y = model.predict(test)
```

```python
#Extract the actual Sale Price value of test data
print("Actual Price of first 5 instances:\n")
print(actual_y.head())
print("\nPredicted Price of first 5 instances:\n")
print(pd.DataFrame(predicted_y).head())
```

```python
%config InlineBackend.figure_format = 'svg'
#Plot the actual Price value against predicted Sale Price 
g = sns.jointplot(x= actual_y, y= predicted_y, kind='reg', xlim = (0,400000), ylim=(0,600000),
                  joint_kws={'line_kws':{'color':'darkorange'}})
g.set_axis_labels(xlabel='Actual Price', ylabel='Predicted Price')
plt.show()
```

<h2>Submission of Results:</h2>

```python
my_submission = pd.DataFrame({'Id': Id, 'SalePrice': predicted_y})

my_submission.to_csv('sample_submission.csv', index=False)
```
